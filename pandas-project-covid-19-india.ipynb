## Asking and Answering Questions

## Question 1

On which day, most number to cases were confirmed in delhi?

```python
# Here we have sorted the values based on the descending order of Total_Confirmed_cases.
# We have used the .head method to retrive the top 5 values.
# The top most value is the most number of cases confirmed.

delhi_df.sort_values('Total_Confirmed_cases', ascending=False).head(5)
```

The most number of confirmed cases were reported on 6th August 2020 in Delhi.

## Question 2

On which day, least number of cases were reported in Delhi?

```python
# Here we have sorted the values based on the descending order of Total_Confirmed_cases.
# We have used the .tail method to retrive the bottom 5 values.
# The bottom most value is the least number of cases confirmed.

delhi_df.sort_values('Total_Confirmed_cases', ascending=False).tail(5)
```

The least number of confirmed cases were reported on 2nd,3rd and 4th of March 2020 in Delhi.

## Question 3

On which day, most number of deaths took place in Delhi?

```python
# Here we have sorted the values based on the descending order of Death.
# We have used the .head method to retrive the top 5 values.
# The top most value is the most number of deaths.

delhi_df.sort_values('Death', ascending=False).head(5)
```

The most number of deaths took place on 11th June 2020 in Delhi.

## Question 4

On which day, most of the people got cured/discharged/migrated in Delhi?

```python
# Here we have sorted the values based on the descending order of Cured/Discharged/Migrated.
# We have used the .head method to retrive the top 5 values.
# The top most value is the most number of people who got Cured/Discharged/Migrated.

delhi_df.sort_values('Cured/Discharged/Migrated', ascending=False).head(5)
```

The most number of people who were cured/discharged/migrated on 6th August 2020 in Delhi.

## Question 5

On which day, most of the people got recovered in Delhi?

```python
# Here we have sorted the values based on the descending order of New_recovered.
# We have used the .head method to retrive the top 5 values.
# The top most value is the most number of people who newly recovered.

delhi_df.sort_values('New_recovered', ascending=False).head(5)
```

Most people recovered on 21st June 2020 in Delhi.

## Question 6

On which day, most number of new cases were reported?

```python
# Here we have sorted the values based on the descending order of New_cases.
# We have used the .head method to retrive the top 5 values.
# The top most value is the most number of newly reported cases.

delhi_df.sort_values('New_cases', ascending=False).head(5)
```

Most number of new cases were reported on 27th June 2020 in Delhi.

## Visualization

```python
# Here we have plotted the Total_Confirmed_cases.

plt.hist(delhi_df.Total_Confirmed_cases);
plt.xlabel('No of cases in thousands')
```

```python
# Here we have plotted the New_cases.

plt.hist(delhi_df.New_cases);
plt.xlabel('No of cases in thousands')
```

```python
# Here we have plotted the New_recovered.

plt.hist(delhi_df.New_recovered);
plt.xlabel('No of cases in thousands')
```

```python
# Here we have plotted the New_cases and New_recovered.

plt.plot(delhi_df.New_cases, delhi_df.New_recovered, 's-b')
plt.xlabel('No of cases in thousands')
plt.ylabel('No of people recovered')
```
